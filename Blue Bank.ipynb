```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_json('loan_data_json.json')
```

```python
loandata = pd.DataFrame(data)
loandata
```

```python

loandata["AnnualIncome"]= np.exp(loandata['log.annual.inc'])
loandata.head(15)
```

```python
def categorize_fico(row):
    try:
        if row < 600:
            val = 'poor'
        elif row >= 600 and row < 680:
            val = 'fair'
        elif row >= 680 and row < 730:
            val = 'good'
        elif row > 730:
            val = 'Excellent'
        else:
            val = 'unknown'
        return val
    except:
        return 'error - unknown'
```

```python
loandata["fico"].describe()
```

```python
# loandata['ficocat'] = loandata.apply(ff, axis=1)

loandata["ficocat"] = loandata["fico"].apply(lambda x: categorize_fico(x))
```

```python
loandata.head(30)
```

```python
loandata.loc[loandata['int.rate']>0.12, 'int_rate_type' ]= "High"
loandata.loc[loandata['int.rate']<=0.12, 'int_rate_type' ]= "Low"   
```

```python
catplot = loandata.groupby(['ficocat']).size()
```

```python
catplot.plot.bar(color = 'orange')
plt.show()
```

```python
purposecount = loandata.groupby(['purpose']).size()
purposecount.plot.bar( )
plt.show()
```
